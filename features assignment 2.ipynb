```python
# ANS 1:
# MinMax scaling
# MinMax is also known as normalization , it is a feature scaling technique used in data preprocessing.
# It transforms the original value of a feature to a new scale between 0-1
# Scaled_value = (Value - Min_value) / (Max_value - Min_value)

# It is used in data preprocessing by which  scaling the features to a common range, 
#  it can help improve the performance of machine learning models. 
```

```python
# ANS 3:
# Principle Component Analysis
# Principle Component Analysis (PCA) is a dimensionality reduction technique used to transform a high-dimensional dataset into a lower-dimensional space while 
#    preserving the most important information or patterns in the data

# It is used for dimentionality redution
# By reducing the dimensionality of the data, it can simplify complex datasets, remove redundant or irrelevant features, 
#  and improve computational efficiency in subsequent analysis or modeling tasks.

# Suppose for example you have a dataset with three features: "Age," "Income," and "Number of Purchases.

# 1.Standardize the data: Subtract the mean and scale the features to have unit variance.
# 2.Compute the covariance matrix: Calculate the covariance matrix based on the standardized data. 
# 3.Compute the eigen vectors and eigenvalues: Find the eigenvectors and eigenvalues of the covariance matrix. 
# 4.Select the principal components: Sort the eigenvectors based on their corresponding eigenvalues in descending order. 
#     Let's say the first two principal components explain 95% of the variance in the data, so we select those two.
# 5 .Transform the data: Project the original data onto the selected principal components. 
#    This will result in a new dataset with two features, which are the lower-dimensional representation of the original data.
```

```python
# ANS 8 :
from sklearn.decomposition import PCA
import numpy as np
data = np.array([
    [160, 60, 25, 0, 120],
    [170, 65, 30, 1, 130],
    [155, 55, 27, 0, 115],
    [180, 70, 35, 1, 140],
    [162, 58, 28, 0, 118]
])

pca = PCA()
pca.fit(data)

cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)

n_components = np.argmax(cumulative_variance_ratio >= 0.95) + 1
print("Cumulative Explained Variance Ratio:", cumulative_variance_ratio)
print("Number of Principal Components to Retain:", n_components)

```

```python
# we retain all five components because we want to retain enough principal components to explain at least 95% of the variance in the data.
```
